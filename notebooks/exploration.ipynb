```python
# Filtrando colunas
dfs = dfs[ ['dataNotificacao', 'cnes', 
              'ocupacaoCovidUti', 'ocupacaoCovidCli', 
              'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli',
              'saidaSuspeitaObitos', 'saidaSuspeitaAltas',
              'saidaConfirmadaObitos', 'saidaConfirmadaAltas', 
              'estado', 'municipio'] ]
              
```

```python
dfs['dataNotificacao'] = pd.DatetimeIndex(pd.to_datetime(dfs['dataNotificacao'],
    format='%Y-%m-%dT%H:%M:%S.%fZ'
    )).normalize()
    
dfs['dataNotificacao']
```

```python
dfs['mes'] = pd.DatetimeIndex(dfs['dataNotificacao']).month
dfs['ano'] = pd.DatetimeIndex(dfs['dataNotificacao']).year
dfs
```

```python
dfs.sort_values(by='dataNotificacao', inplace = True)
```

```python
def get_region(state):
    """state deve ser uma string com o nome do estado"""

    if state in ['Goiás', 'Mato Grosso', 'Mato Grosso do Sul', 'Distrito Federal']:
        return "Centro-Oeste"
    elif state in ['Acre', 'Amazonas', 'Amapá', 'Pará', 'Rondônia', 'Roraima', 'Tocantins']:
        return "Norte"
    elif state in ['Alagoas', 'Bahia', 'Ceará', 'Maranhão', 'Piauí', 'Pernambuco', 'Paraíba', 'Rio Grande do Norte', 'Sergipe']:
        return "Nordeste"
    elif state in ['Paraná', 'Rio Grande do Sul', 'Santa Catarina']:
        return "Sul"
    elif state in ['Espírito Santo', 'Minas Gerais', 'Rio de Janeiro', 'São Paulo']:
        return "Sudeste"
    else:
        return "NA"

# Criando região e mes 
dfs['regiao'] = dfs['estado'].apply(get_region)
dfs
```

```python
# Calculando média móvel nos dados

for col in ['ocupacaoCovidUti', 'ocupacaoCovidCli', 
              'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli',
              'saidaSuspeitaObitos', 'saidaSuspeitaAltas',
              'saidaConfirmadaObitos', 'saidaConfirmadaAltas']:

    dfs[f'{col}_movingAverage'] = dfs[col].rolling(window=7).mean()
```

```python
output_file("../visualizations/vis1.html")

dfs_g = dfs.groupby(by='dataNotificacao').sum(numeric_only=True).reset_index()

source = ColumnDataSource(data={
    "date" : dfs_g['dataNotificacao'],

    "Ocupação de UTI"                : dfs_g['ocupacaoCovidUti_movingAverage'],
    "Ocupação de clínicas"           : dfs_g['ocupacaoCovidCli_movingAverage'],
    "Óbitos decorrentes de COVID-19" : dfs_g['saidaConfirmadaObitos_movingAverage'],
    "Óbitos de suspeita de COVID-19" : dfs_g['saidaSuspeitaObitos_movingAverage'],
    "Altas de COVID-19"              : dfs_g['saidaConfirmadaAltas_movingAverage'],
    "Altas de suspeita de COVID-19"  : dfs_g['saidaSuspeitaAltas_movingAverage'],
})

faixaDatas = np.array(dfs_g['dataNotificacao'], dtype=np.datetime64)

TOOLS = "xpan,wheel_zoom,box_zoom,reset,save,tap" #pan

plot = figure(
    height=500, width=750, tools=TOOLS, toolbar_location='right',
    title="Médias Móveis de Indicadores do SUS de ocupação, óbito e recuperação por COVID-19",
    x_axis_type="datetime", x_axis_location="above",
    background_fill_color="#efefef", x_range=(faixaDatas[0], 
    faixaDatas[len(faixaDatas)-1])
)
plot.yaxis.axis_label = 'Pessoas'
    
# Configura BOX de seleção para detalhar porções do comportamento da curva de cotações
select = figure(title="Ajuste o box de seleção para ver detalhes",
                height=100, width=750, y_range=plot.y_range,
                x_axis_type="datetime", y_axis_type=None,
                tools="", toolbar_location=None, background_fill_color="#efefef")

plot.add_tools(HoverTool(
    tooltips=[
        ( "Data",  "$x{%F}"            ),
        ( "Valor", "$y{"+f"0.00"+" a}" ),
    ],

    formatters={
        '$x'      : 'datetime', # use 'datetime' formatter for 'date' field
    },

    # display a tooltip whenever the cursor is vertically in line with a glyph
#     mode='vline'
))

range_tool = RangeTool(x_range=plot.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.3

select.ygrid.grid_line_color = None
select.add_tools(range_tool)

# Set up MultiChoice widget
options = ['Ocupação de UTI', 'Ocupação de clínicas', 'Óbitos decorrentes de COVID-19',
          'Óbitos de suspeita de COVID-19', 'Altas de COVID-19', 'Altas de suspeita de COVID-19']
colors  = ['red', 'green', 'blue', 'cyan', 'purple', 'gray']
multi_choice = MultiChoice(value=options[:3], options=options, max_items=3, title='Seleção:')

name_dict={'name':[],'minimap':[],'legend':[],'label':[]}
for name, color in zip(options, colors):  
    name_glyph    = plot.line('date', name, source=source, legend_label=name, color=color)
    minimap_glyph = select.line('date', name, source=source, color=color)
    
    name_dict['name'   ].append(name_glyph)
    name_dict['minimap'].append(minimap_glyph)
    name_dict['label'  ].append(name)

name_legend_dict={'name':[]}
for label in range(len(options)):
    name_dict['legend'].append(plot.legend.items[label])

for i in range(len(options)):
    if name_dict['label'][i] in options[:3]:
        name_dict['name'][i].visible = True;
        name_dict['minimap'][i].visible = True;
        name_dict['legend'][i].visible = True;
    else:
        name_dict['name'][i].visible = False;
        name_dict['minimap'][i].visible = False;
        name_dict['legend'][i].visible = False;   
    
callback = CustomJS(args=dict(name_dict=name_dict, multi_choice=multi_choice), code="""
var selected_vals = multi_choice.value;
var index_check = [];

for (var i = 0; i < name_dict['name'].length; i++) {
    index_check[i]=selected_vals.indexOf(name_dict['label'][i]);
        if ((index_check[i])>= 0) {
            name_dict['name'][i].visible = true;
            name_dict['minimap'][i].visible = true;
            name_dict['legend'][i].visible = true;
            }
        else {
            name_dict['name'][i].visible = false;
            name_dict['minimap'][i].visible = false;
            name_dict['legend'][i].visible = false;
        }
    }
""")

multi_choice.js_on_change('value', callback)

plot.legend.background_fill_alpha = 0.25

show(column(multi_choice, plot, select))
```

```python
output_file("../visualizations/vis3.html")

x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create three plots

dfs_g = dfs.groupby(by=['ano', 'mes']).sum(numeric_only=True).reset_index()
dfs_g['saidaObitosPercentage'] = 100* dfs_g['saidaConfirmadaObitos'] / (dfs_g['saidaConfirmadaAltas']+dfs_g['saidaConfirmadaObitos'])
    
plots = []
for col, color, title in [
    ('ocupacaoCovidUti',      'purple',  'Ocupação de UTI por pacientes com COVID-19'),
    ('ocupacaoHospitalarUti', 'cyan', 'Ocupação de UTI Hospitalar por pacientes com COVID-19'),
    ('saidaConfirmadaAltas', 'green', 'Altas de pacientes com COVID-19'),
    ('saidaConfirmadaObitos', 'red', 'Óbitos de pacientes com COVID-19'),
    ('saidaObitosPercentage', 'black', 'Porcentagem de Óbitos nas saídas')
]:

    dfs_altas = dfs_g.pivot(index='ano', columns='mes', values=col)
    source_altas = ColumnDataSource(data={
        'mes'  : [str(i) for i in range(1, 13)],
        'data' : dfs_altas.loc[2022, :].values
    })
    s = figure(x_range=[str(i) for i in range(1, 13)], y_range=( (0, 4e+5) if col!='saidaObitosPercentage' else (0, 100)),
               width=750, height=250, title=title,
               toolbar_location=None, tools="")
    
    s.vbar(x='mes', top='data', width=0.9, source=source_altas, color=color)
    
    s.yaxis.axis_label = ('Pessoas internadas' if col!='saidaObitosPercentage' else "Porcentagem")

    s.add_tools(HoverTool(
    tooltips=[
        ("Mês",           "@mes"),
        ("Pessoas:",        "@data")
    ],

        # display a tooltip whenever the cursor is vertically in line with a glyph
        #mode='vline'
    ))
    plots.append(s)

s.xaxis.axis_label = 'Mês do ano'
# put the results in a column and show
show(column( *plots ))
```
